# ðŸŒ¸ Task 1: Exploring and Visualizing the Iris Dataset

This notebook explores the classic **Iris flower dataset** using pandas, seaborn, and matplotlib.  
We will perform:
- Data loading and inspection
- Summary statistics
- Missing value checks
- Scatter plot, histogram, and box plot visualizations


```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
```

## ðŸ“¥ Load the Iris Dataset

We'll load the built-in Iris dataset from seaborn and inspect it.


```python
# Load Iris dataset
iris = sns.load_dataset("iris")

# Show shape, columns, and first few rows
print("Shape of dataset:", iris.shape)
print("Columns:", iris.columns.tolist())
iris.head()
```

```python
# Summary statistics
iris.describe()
```

```python
# Check for missing values
iris.isnull().sum()
```

## ðŸ“Š Scatter Plot: Petal Length vs Petal Width

This plot helps us see how petal dimensions vary across species.


```python
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species", palette="bright")
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title="Species")
plt.show()
```

```python
sns.histplot(data=iris, x="sepal_length", kde=True, color="skyblue", bins=15)
plt.title("Histogram: Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(data=iris, x="species", y="sepal_width", palette="Set2")
plt.title("Box Plot: Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```
